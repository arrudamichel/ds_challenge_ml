## 3. Merge em dataframe único

### 3.1. Merge dos datasets

```python
df_features["Date"] = pd.to_datetime(df_features["Date"], dayfirst=True)
df_sales["Date"] = pd.to_datetime(df_sales["Date"], dayfirst=True)
```

```python
# Merge entre vendas e características (combinando por Store, Date, IsHoliday)
df = df_sales.merge(df_features, on=["Store", "Date", "IsHoliday"], how="left")

# Merge com informações das lojas
df = df.merge(df_stores, on="Store", how="left")
```

```python
df_unemployment = df.copy()
df_unemployment = df_unemployment.sort_values('Date')

grouped = df_unemployment.groupby(['Date', 'Store'])['Unemployment'].mean().reset_index()

plt.figure(figsize=(10, 6))
for store in grouped['Store'].unique():
    store_data = grouped[grouped['Store'] == store]
    plt.plot(store_data['Date'], store_data['Unemployment'], label=f'Loja {store}')

# Títulos e rótulos
plt.title('Desemprego médio ao longo do tempo por loja', fontsize=14)
plt.xlabel('Data', fontsize=10)
plt.ylabel('Desemprego médio (%)', fontsize=10)
plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```

##### Possíveis outliers? 
- Weekly_Sales: Faz sentido terem outliers por causa das sazonalidades
- Markdown1, Markdown2, Markdown3, Markdown4, Markdown5: Necessário entender a relação entre a sazonalidade e os descontos
- IsHoliday: Só possui valores True ou False
- Temperature: Aparentemente, os valores -2.06 estão na mesma região e a temperatura nessa região pode variar entre -2.06 a 68.84. Levando em consideração que os feriados são norte americanos, essas medidas devem ser em fahrenheit. Convertendo para celsius, ficaria entre -18 graus até 20 graus censius. Realizei uma breve pesquisa e cheguei a conclusão de que existem regiões do norte dos EUA que tem a possibilidade de ter essa variação de temperatura, como Minnesota ou Montana.
- Unemployment: Acredito que os valores discrepantes estão no mesmo período do tempo. Parece interessante ver a relação do desemprego com as vendas e descontos.

## 7. Pré-processamento dos Dados

### 7.1. Preencher descontos

```python
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
df[markdown_cols] = df[markdown_cols].fillna(0)
```

### 7.2. Criar features de data

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.isocalendar().week.astype(int)
df['DayOfYear'] = df['Date'].dt.dayofyear
```

### 7.3. Codificar type para analises

```python
df["Type"] = df["Type"].map({'A': 0, 'B': 1, 'C': 2})
```

### 7.4. Criar colunas a partir de desconto

```python
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
```

#### 7.4.1 Soma de descontos na compra

```python
df['DescontoSum'] = df[markdown_cols].sum(axis=1)
```

```python
df['DescontoSum']
```

#### 7.4.2 Tem de descontos na compra?

```python
df['TemDesconto'] = df['DescontoSum'] > 0
```

```python
df['TemDesconto']
```

#### 7.4.3 Media de descontos

```python
df['DescontoAvg'] = df[markdown_cols].mean(axis=1)
```
